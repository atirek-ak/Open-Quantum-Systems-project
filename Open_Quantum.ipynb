```
import matplotlib.pyplot as plt
import numpy as np
import math
```

```
#eigenvalue and eigenvector
m = np.array([[1, 2], [2, 3]]) 
w, v = np.linalg.eig(m) 
print("Printing the Eigen values of the given square array:\n", w) 
print("Printing Right eigenvectors of the given square array:\n", v)
```

```
#plot for free space
t = np.linspace(0,100,num=300)
# t = np.linspace(np.log(2),(5*np.pi)/6,num=10)
y1 = abs(np.sqrt(1 - 2*np.exp(-t)))
y2 = abs(np.sqrt(1 - 2*np.cos(t)))
# fig = plt.figure()
plt.figure(figsize=(12,7))
plt.plot(t,y1, 'r')
plt.plot(t,y2, 'b')
# plt.margins(x=-0.49)
plt.legend(["exp function", "cos function"], loc ="lower right") 
plt.show()
```

```
#plot for constrained space
# t = np.linspace(0,100,num=300)
t = np.linspace(np.log(2),(5*np.pi)/6,num=10)
y1 = abs(np.sqrt(1 - 2*np.exp(-t)))
y2 = abs(np.sqrt(1 - 2*np.cos(t)))
# fig = plt.figure()
plt.figure(figsize=(12,7))
plt.plot(t,y1, 'r')
plt.plot(t,y2, 'b')
plt.legend(["exp function", "cos function"], loc ="lower right") 
# plt.margins(x=-0.49)
plt.show()
```
